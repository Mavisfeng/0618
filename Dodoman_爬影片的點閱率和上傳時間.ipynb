```python
from googleapiclient.discovery import build

# 設置 YouTube API 金鑰
API_KEY = 'AIzaSyAjkvZktxgHI0-6LPZ0QYQre_f15Q_vlGk'  # 替換為你的 YouTube Data API 金鑰
YOUTUBE_CHANNEL_ID = 'UCfq75-6J5seC82CmtLSFxXw'  # 替換為目標 YouTube 頻道 ID

# 初始化 YouTube API
youtube = build('youtube', 'v3', developerKey=API_KEY)

# 獲取特定 Youtuber 的頻道 ID
def get_channel_id(channel_name):
    search_response = youtube.search().list(
        q=channel_name,
        part='snippet',
        type='channel',
        maxResults=1
    ).execute()
    return search_response['items'][0]['id']['channelId']

# 輸入你想搜尋的 Youtuber 名字
channel_name = "The DoDo Men - 嘟嘟人"
channel_id = get_channel_id(channel_name)

# 獲取頻道的影片資訊
def get_videos_from_channel(channel_id):
    videos = []
    request = youtube.search().list(
        part='snippet',
        channelId=channel_id,
        maxResults=50,  # 每次最多取得 50 筆資料
        order='date'  # 按時間排序
    )
    response = request.execute()

    # 迴圈處理影片清單
    for item in response['items']:
        if item['id']['kind'] == 'youtube#video':
            video_id = item['id']['videoId']
            title = item['snippet']['title']
            publish_time = item['snippet']['publishedAt']

            # 獲取影片詳細資訊（包括點閱率）
            video_details = youtube.videos().list(
                part='statistics',
                id=video_id
            ).execute()
            view_count = video_details['items'][0]['statistics'].get('viewCount', 0)

            videos.append({
                'title': title,
                'publish_time': publish_time,
                'view_count': view_count
            })
    return videos

# 爬取目標頻道的影片
videos = get_videos_from_channel(YOUTUBE_CHANNEL_ID)

# 將結果顯示出來
for video in videos:
    print(f"Title: {video['title']}")
    print(f"Published At: {video['publish_time']}")
    print(f"View Count: {video['view_count']}")
    print('-' * 50)
```
